
# Handle missing values
---
### There are a few options for handling missing values:

    - Remove any rows that have missing values.
    - Remove any columns that have missing values.
    - Fill the missing values with some other value.
    - Leave the missing values as is.

The first two options are often used to prepare data for machine learning algorithms, which are unable to be used with data that includes null values. We can use the **DataFrame.dropna()** method to remove or drop rows and columns with null values.

Syntax:

#### DataFrame.dropna(axis=0, how='any',  inplace=False)

Arguments:
###### axis: Default â€“ 0
   -  0, or â€˜indexâ€™ : Drop rows which contain NaN values.
   - 1, or â€˜columnsâ€™ : Drop columns which contain NaN value.
    
###### how: Default â€“ â€˜anyâ€™
   - â€˜anyâ€™ : Drop rows / columns which contain any NaN values.
   - â€˜allâ€™ : Drop rows / columns which contain all NaN values.
    
###### inplace (bool): Default- False
   - If True, modifies the calling dataframe object

```python
# Number of missing values in each coulmn
df1.isnull().sum()
```

### Dropping the rows which has missing value in calorie column

```python
df1.dropna(subset=['calories'], inplace=True)
```

```python
df1.isnull().sum()
```

#### Fill the missing values with some other value. In this case it has been filled with zero.


```python
df1 = df1.fillna(0)
```

```python
df1.isnull().sum()
```

```python
df1.info()
```

```python
# Reset the index of the DataFrame
df1.reset_index(drop=True, inplace = True)
df1
```
